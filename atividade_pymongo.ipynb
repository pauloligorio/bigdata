```python
#Impotando pacotes
import pymongo
from pymongo import MongoClient
```

```python
#Conectando ao banco de dados com a chave de conexão
client = pymongo.MongoClient("mongodb+srv://admin:admin@cluster0.fa9f4.gcp.mongodb.net/anatel?retryWrites=true&w=majority")
db = client.anatel
```

```python
#1 - Quantos documentos possui a coleção minas?
db.minas.count_documents({})
```

```python
#2 - Mostre os 5 primeiros documentos da coleção.
list(db.minas.find().limit(5))
```

```python
#3 - Selecione o documento que possui o maior número de solicitações QtdeSolic na Anatel.
db.minas.find_one(sort=[("QtdeSolic", -1)])
```

```python
#4 - Qual é o menor número de solicitações QtdeSolic na Anatel?
foo = db.minas.find_one(sort=[("QtdeSolic", 1)])
print(foo['QtdeSolic'])
```

```python
#5 - Mostre todos os documentos com o menor número de solicitações QtdeSolic na Anatel.

var = db.minas.find_one(sort=[("QtdeSolic", 1)])
menorQtde = var['QtdeSolic']
list(db.minas.find({'QtdeSolic': menorQtde}))
```

```python
#6 - Mostre as contagens de solicitações QtdeSolic por tipo de serviço Servico, em ordem decrescente pelas contagens
pipeline = [
    { "$group": { "_id": "$Servico", "contagem": { "$sum": "$QtdeSolic"  } } },
    { "$sort": { "contagem": -1 }}
]

list(db.minas.aggregate(pipeline))
```

```python
#7 - Mostre todos os documentos com atributo Tipo igual a "Reclamação".
reclamacoes = list(db.minas.find({ "Tipo": "Reclamação"}))
len(reclamacoes)

```

```python
#8 - Quais diferentes valores encontramos em Tipo?
list(db.minas.distinct('Tipo'))
```
